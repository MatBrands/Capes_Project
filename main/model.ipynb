```python
import pandas as pd
import numpy as np
from nltk import FreqDist
from nltk.tokenize import wordpunct_tokenize
```

```python
year = 2013
df = pd.read_pickle(f'./Processed/{year}_stem.pkl')
```

```python
test = list(set([item for item in wordpunct_tokenize(' '.join(df.resumo)) if len(item) > 2]))
```

```python
len(test)
```

```python
freq = dict()

for character in test:
    freq[character] = {}
    for item in sorted(df.area_avaliacao.unique()):
        freq[character][item] = 0
        
    areas = df[df.resumo.str.contains(character)].area_avaliacao
    for item in sorted(areas.unique()):
        freq[character][item] = areas.value_counts()[item]
        
freq = pd.DataFrame(freq).transpose()
```
