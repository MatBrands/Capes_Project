```python
import pandas as pd
import numpy as np
from nltk import FreqDist
from nltk.tokenize import wordpunct_tokenize
```

```python
year = 2013
df = pd.read_pickle(f'./Processed/{year}_stem.pkl')
```

```python
df.head()
```

```python
def freq_dist(array: list) -> dict:
    text = ' '.join(array)
    return FreqDist(wordpunct_tokenize(text))
```

```python
frequencia = freq_dist(df.resumo)

frequencia.most_common(5)
```

```python
freq = dict(frequencia)

for character in freq:
    aux = freq[character]
    freq[character] = {}
    freq[character]['frequencia'] = aux
    for item in sorted(df.area_avaliacao.unique()):
        freq[character][item] = 0
        
    areas = df[df.resumo.str.contains(character)].area_avaliacao
    for item in sorted(areas.unique()):
        freq[character][item] = areas[areas.str.contains(item)].shape[0]
        
freq = pd.DataFrame(freq).transpose()
```

```python
freq.sort_values(by='frequencia', ascending=False)
```
